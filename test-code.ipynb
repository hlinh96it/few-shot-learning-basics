```python
import os
import glob
import pandas as pd
from pathlib import Path
import shutil
```

### Data preprocessing

```python
dataset_path = '/Users/hoanglinh96nl/FashionNet Dataset/myntradataset/'
label = pd.read_csv('/Users/hoanglinh96nl/FashionNet Dataset/myntradataset/styles.csv', on_bad_lines='skip')
label
```

```python
label['masterCategory'].value_counts()
```

```python
df_group = label.groupby('masterCategory')

df_2000 = pd.DataFrame()
for key, group in df_group:
    try:
        sample_2000 = group.sample(n=2000)
        df_2000 = pd.concat([df_2000, sample_2000])
    except ValueError:
        pass
```

```python
df_2000
```

```python
for img_index, row in df_2000.iterrows():
    image_name = Path(dataset_path, 'images', str(row['id']) + '.jpg')
    class_name = row['masterCategory']
    target_path = Path(dataset_path, class_name)
    if not Path.is_dir(Path(dataset_path, class_name)):
        target_path.mkdir()
        
    try:
        shutil.move(image_name, target_path)
    except:
        pass
```
